#### NO HAY NULOS
En este caso no hay valores nulos.  
Con carácter general los nulos se pueden acometer de dos maneras:
- Eliminando la columna con nulos. Si la columna en cuestión tiene un número de nulos muy alto en relación con el número de filas del DataFrame, la columna entera podría ser irrelevante. En ese caso se eliminaría directamente.
Esto se hará de la siguiente forma:
<pre>df.drop(["col1", "col2", "col3"], inplace=True)</pre>  


- Cumplimentando los valores nulos, si estos no son una cantidad importante, con valores por defecto. Para ello se pueden adoptar diversos criterios, según el tipo de dato de una columna en concreto. Los más habituales son:
    - Para columnas numéricas, emplear la media de los valores no nulos de la propia columna.
    - Para columnas de tipo string, un valor tipo, como "Unknown" o "Desconocido"
    - Para columnas booleanas, el valor False.
Esto se hará de la siguiente forma:
<pre>def fill_na(df):
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].mean(), inplace=True)
        elif pd.api.types.is_bool_dtype(df[col]):
            df[col].fillna(False, inplace=True)
        else:
            df[col].fillna("Unknown", inplace=True)
    return df<pre>

<pre>df = fill_na(df)</pre>
